```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import seaborn as sns
from datetime import datetime
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

sns.set_style = "darkgrid"

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
dct = {}
cols = ["City","Lat","Lng","Max Temp","Humidity","Cloudiness","Wind Speed","Country","Date"]
for city in cities:
    city_weather = requests.get(f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={weather_api_key}').json()
    #print(city_weather.json())
    if city_weather["cod"] == 200:
        dct[city_weather["id"]] = []
        dct[city_weather["id"]].append(city_weather["name"])
        dct[city_weather["id"]].append(city_weather["coord"]["lat"])
        dct[city_weather["id"]].append(city_weather["coord"]["lon"])
        dct[city_weather["id"]].append(city_weather["main"]["temp_max"])
        dct[city_weather["id"]].append(city_weather["main"]["humidity"])
        dct[city_weather["id"]].append(city_weather["clouds"]["all"])
        dct[city_weather["id"]].append(city_weather["wind"]["speed"])
        dct[city_weather["id"]].append(city_weather["sys"]["country"])
        dct[city_weather["id"]].append(datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S'))
        print(f'Read in {city_weather["id"]} - {city}')
df = pd.DataFrame.from_dict(dct, orient="index", columns=cols)
```

### Convert Raw Data to DataFrame

```python
axes_labels = {'City':'City', 'Lat':'Latitude', 'Lng':'Longitude', 'Max Temp':'Maximum Temperature in Degrees Kelvin', 'Humidity':'Humidity',
 'Cloudiness':'Cloudiness', 'Wind Speed':'Wind Speed', 'Country':'Country', 'Date':'Date'}
def plot_scatter(df,x,y,title, save=True):
    fig, ax = plt.subplots()
    ax.scatter(df[x],df[y])
    ax.xlabel = axes_labels[x]
    ax.ylabel = axes_labels[y]
    plt.title(title+" on 2/12/2021");
    if save:
        plt.savefig(f"{title}.png")
    
```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
df_north = df[df["Lat"]>=0]
slope, intercept, rvalue, pvalue, stderr = linregress(df["Max Temp"],df["Lat"])
print(f"slope = {slope}, intercept = {intercept}, rvalue = {rvalue}, pvalue = {pvalue}, stderr = {stderr}")
plot_scatter(df_north,"Max Temp","Lat", "Max Temp vs Latitude")
x = df_north["Max Temp"]
plt.plot(x, (x*slope)+intercept, color="r")
```

```python
def plot_scatter_linreg(df,x,y,title, save=True):
    slope, intercept, rvalue, pvalue, stderr = linregress(df[x],df[y])
    plot_scatter(df,x,y, title, False)
    x = df[x]
    plt.plot(x, (x*slope)+intercept, color="r");
    print(f"r = {rvalue}")
    if save:
        plt.savefig(f"{title}.png")
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
df_south = df[df["Lat"]<=0]
plot_scatter_linreg(df_south,"Max Temp","Lat","Max Temp vs Latitude in the Southern Hemisphere")
```
